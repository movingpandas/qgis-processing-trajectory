```python
tc= mpd.TrajectoryCollection(gdf, traj_id_col="trajectory_id", t="t")
```

```python
tc.add_direction()
```

```python
gdf = tc.to_point_gdf()
gdf
```

```python
gdf["t"] = gdf.index
```

```python
gdf
```

```python
from qgis.PyQt.QtCore import QVariant
from qgis.core import QgsField
timestamp_field = "t"
                       
output_fields = vl.fields()

output_fields.append(QgsField(tc.get_direction_col(), QVariant.Double))
i = output_fields.indexFromName("fid")
output_fields.remove(i)
i = output_fields.indexFromName(timestamp_field)
output_fields.remove(i)
output_fields.append(QgsField(timestamp_field, QVariant.String))

for field in  vl.fields():
    print(field.name())
```

```python
names = [f.name() for f in output_fields]
names.append('geometry')
gdf[names]
```

```python
gdf[timestamp_field] = gdf.index.astype(str)

for _, row in gdf[names].iterrows():
    print(row.values.tolist()[:-1])
```
